## 2 . Identify all integers whose digit sum is a prime number.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True
is_prime(2)
```

## 5 . Determine all numbers where the value equals the sum of its index and the reverse of the index

```python
def numbers(numbers):
    result = []
    for i, n in enumerate(numbers):
        temp = i
        rev = 0
        while temp > 0:
            rev = rev * 10 + temp % 10
            temp //= 10
        if n == i + rev:
            result += [n]
    return result
nums = [0, 1, 3, 12, 22, 44, 55]
numbers(nums)
```

## 7. Identify numbers that remain the same when reversed and are divisible by 11.

```python
def div11(numbers):
    result = []
    for n in numbers:
        temp = n
        rev = 0
        while temp > 0:
            rev = rev * 10 + temp % 10
            temp //= 10
        if n == rev and n % 11 == 0:
            result += [n]  
    return result
nums = [11, 22, 121, 131, 242, 44]
div11(nums)
```

## 8. Determine the product of all non-zero digits of an integer

```python
def product(n):
    if n < 0:
        n = -n  
    product = 1
    if n == 0:  
        return 0
    while n > 0:
        digit = n % 10
        if digit != 0:
            product *= digit
        n //= 10
    return product
nums = [10, 20, 90, 0, -204]
for num in nums:
    print(product(num))
```

## 11. Find all integers where the product of digits is greater than the sum of digits.

```python
def product(numbers):
    result = []
    for n in numbers:
        temp = n
        s = 0
        p = 1
        while temp > 0:
            d = temp % 10
            s += d
            p *= d
            temp //= 10
        if p > s:
            result += [n]
    return result
nums = [12, 13, 22, 23, 34]
print(product(nums))
```

## 12. For each number, compute the count of binary 1 digits without using direct built-in binary conversion

 ## 13. Check whether the sum of digits of an integer is equal to its reverse

```python
def reverse(n):
    temp = n
    s = 0
    while temp > 0:
        s += temp % 10
        temp //= 10
    rev = 0
    temp = s
    while temp > 0:
        rev = rev * 10 + temp % 10
        temp //= 10
    return s == rev
nums = [12, 21, 5, 123,]
for num in nums:
    if reverse(num):
        print(num)
```

 ## 20. Identify numbers divisible by the sum of their digits and whose reversed form is an even number

```python
def reverse_even(start, end):
    result = []
    for n in range(start, end + 1):
        sum_digits = 0
        temp = n
        while temp > 0:
            sum_digits += temp % 10
            temp //= 10
        if n % sum_digits != 0:
            continue
        rev = 0
        temp = n
        while temp > 0:
            rev = rev * 10 + temp % 10
            temp //= 10

        if rev % 2 == 0:
            result = result + [n]  
    return result
print(reverse_even(1, 25))
```

## 21. Convert each integer into the result of adding its first and last digit multiplied by the number of digits

```python
def number(n):
    last = n % 10
    first = n
    digits = 0
    while first >= 10:
        first //= 10
    temp = n
    while temp > 0:
        digits += 1
        temp //= 10
    return (first + last) * digits
numbers = [123, 45, 7, 890]
for n in numbers:
    print(number(n))
```

 ## 26. Convert integers into the difference between the sum of even digits and sum of odd digits.

```python
def even_odd_diff(n):
    diff = 0
    while n > 0:
        digit = n % 10
        if digit % 2 == 0:
            diff += digit
        else:
            diff -= digit
        n //= 10
    return diff
numbers = [123, 456, 789, 2468]
for num in numbers:
    print(even_odd_diff(num))
```

## 27. From a list of integers, keep only those whose middle digit is prime (for odd-length numbers).

```python
def has_prime_middle(n):
    numbers = [123, 456, 789, 135, 246, 97531]   
    temp = n
    digits = 0
    while temp > 0:
        temp //= 10
        digits += 1
    if digits % 2 == 0:
        return False
    temp = n
    for _ in range(digits // 2):
        temp //= 10
    mid_digit = temp % 10
    return mid_digit in [2, 3, 5, 7]
for n in numbers:
    if has_prime_middle(n):
        print(n)
```

 ## 28. Rearrange digits of a number to form the largest possible integer without using sorting utilities.

```python
def largest_number(n):
    count = [0]*10
    temp = n
    while temp > 0:
        count[temp % 10] += 1
        temp //= 10
    result = 0
    for digit in range(9, -1, -1):
        while count[digit] > 0:
            result = result * 10 + digit
            count[digit] -= 1
    return result
numbers = [4213, 305, 987, 1204]
for n in numbers:
    print(largest_number(n))
```

 ## 30. Find all integers for which the index divided by the value yields a whole number

```python
def index_divisible(lst):
    n = len(lst)
    result = [0]*n
    count = 0
    for i in range(n):
        if lst[i] != 0 and i % lst[i] == 0:
            result[count] = lst[i]
            count += 1
    return result[:count]
numbers = [2, 3, 1, 4, 0, 5]
print(index_divisible(numbers))
```

 ## 31. Extract numbers that equal the sum of squares of their digits plus 10

```python
def sum_squares(numbers):
    result = []
    for n in numbers:
        temp = n
        s = 0
        while temp > 0:
            d = temp % 10
            s += d * d
            temp //= 10
        if s + 10 == n:
            result += [n] 
    return result
nums = [19, 29, 30, 91, 50, 10]
print(sum_squares(nums))
```

 ## 32. Convert each number into a running alternating digit sum pattern: d1 - d2 + d3 - d4 and so on

```python
def alternating(n):
    n = abs(n)
    total = 0
    sign = 1
    while n > 0:
        digit = n % 10
        total = digit * sign + total
        sign *= -1 
        n //= 10
    return total
numbers = [1234, 567, 8901]
result = [alternating(num) for num in numbers]
print(result)
```

 ## 33. Find the maximum difference between the product of digits and sum of digits across a list.
 

```python
def max_digit(numbers):
    max_diff = -float('inf')
    for n in numbers:
        prod, summ = 1, 0
        temp = n
        while temp:
            d = temp % 10
            prod *= d
            summ += d
            temp //= 10
        if prod - summ > max_diff:
            max_diff = prod - summ
    return max_diff
numbers = [123, 45, 56, 9]
print(max_digit(numbers))
```

```python
def index_cube(lst):
    result = []
    i = 0
    while i < len(lst):
        n = lst[i]
        if n < 0:
            n = -n  
        total = 0
        temp = n
        while temp > 0:
            total += temp % 10
            temp //= 10
        if i * i * i == total:  
            result += [lst[i]]
        i += 1
    return result
numbers = [0, 1, 8, 9, 17]
print(index_cube(numbers))
```

## 37. Identify numbers where the difference between largest and smallest digit divides the number exactly

```python
def number(numbers):
    result = []
    for n in numbers:
        temp = n
        largest = 0
        smallest = 9
        while temp > 0:
            d = temp % 10
            if d > largest:
                largest = d
            if d < smallest:
                smallest = d
            temp //= 10
        diff = largest - smallest
        if diff != 0 and n % diff == 0:
            result += [n]
    return result
nums = [12, 24, 36, 48]
print(number(nums))
```

 ## 38. Determine the digit that appears most frequently inside a long integer.


```python
def digit(n):
    count = [0] * 10 
    num = n
    if num < 0:
        num = -num  
    while num > 0:
        digit = num % 10
        count[digit] += 1
        num //= 10
number = -1223334
print(digit(number))  
```
