```python
 #1. sum_of_coprimes(n): Return the sum of all numbers less than n that are coprime with n.
# Example: n=10 → coprime numbers = [1,3,7,9] → sum = 20. Hint: Use gcd() logic or check
 #manually for each number.
def sum_of_coprimes(n):
    sum = 0
    for i in range(1, n):
        a = n
        b = i
        while b != 0:
            a, b = b, a % b
        if a == 1:
            sum += i
    return sum
sum_of_coprimes(10)
```

```python
#2. count_armstrong_in_range(a, b): Count how many Armstrong numbers exist between a and b.
 #Example: range(100, 200) → 153 is Armstrong → count=1. Hint: Sum of each digit raised to power
 #of total digits should equal the number.
def count_armstrong_in_range(a, b):
    count = 0
    for num in range(a, b + 1):
        digits = str(num)
        power = len(digits)
        total = 0
        for d in digits:
            total += int(d) ** power
        if total == num:
            count += 1
    return count
count_armstrong_in_range(100, 200)
```

```python
#3. is_krishnamurthy(n): Check if the sum of factorials of digits equals the number. Example: 145
# → 1!+4!+5! = 145 → True. Hint: Use a loop to calculate factorial of each digit.
def is_krishnamurthy(n):
    temp = n
    sum = 0
    while temp > 0:
        digit = temp % 10
        fact = 1
        for i in range(1, digit + 1):
            fact *= i
        sum += fact
        temp //= 10
    return sum == n
is_krishnamurthy(145)

```

```python
#4. count_digits_sum_even(n): Count digits that are even in the number. Example: 1234 → even
# digits = 2,4 → count=2. Hint: Use modulus (%) and division (//) to extract digits.
def count_digits_sum_even(n):
    count = 0
    while n > 0:
        digit = n % 10
        if digit % 2 == 0:
            count += 1
        n //= 10
    return count
count_digits_sum_even(1234)
```

```python
#5. sum_of_perfect_squares(limit): Return sum of all perfect squares up to the limit. Example:
# limit=10 → 1+4+9=14. Hint: Use (i*i) ≤ limit in a loop.
def sum_of_perfect_squares(limit):
    sum = 0
    i = 1
    while i * i <= limit:
        sum += i * i
        i += 1
    return sum
sum_of_perfect_squares(10)
```

```python
# 8. sum_of_even_factors(n): Return sum of even factors of n. Example: n=12 → even
# factors=2,4,6,12 → sum=24. Hint: Check divisibility and parity of factors.
def sum_of_even_factors(n):
    total = 0
    for i in range(1, n + 1):
        if n % i == 0 and i % 2 == 0:
            total += i
    return total
sum_of_even_factors(12)
```

```python
#10. common_factors(a, b): Return all common factors of two numbers. Example: a=12, b=18 →
# [1,2,3,6]. Hint: Check i divides both a and b.
def common_factors(a, b):
    factors = []
    for i in range(1, min(a, b) + 1):
        if a % i == 0 and b % i == 0:
            factors += [i]  
    return factors
common_factors(12, 18)

```

```python
 #11. sum_of_unique_digits(n): Return sum of digits appearing only once. Example: 122334 →
# unique digits=1,4 → sum=5. Hint: Use count() to find unique ones.
def sum_of_unique_digits(n):
    s = str(n)
    sum1 = 0
    for d in s:
        if s.count(d) == 1:
            sum1 += int(d)
    return sum1
sum_of_unique_digits(122334)
```

```python
#12. swap_first_last_digits(n): Swap first and last digits of number. Example: 12345 → 52341.
 #Hint: Convert to string or use integer math with powers of 10.
def swap_first_last_digits(n):
    s = str(n)
    if len(s) == 1:
        return n
    return int(s[-1] + s[1:-1] + s[0])
swap_first_last_digits(12345)
```

```python
#13. is_palindromic_sum(n): Check if the sum of digits is a palindrome. Example: 1234 → sum=10
# → not palindrome. Hint: Find sum, convert to string and compare reversed version.
def is_palindromic_sum(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    s = str(total)
    return s == s[-1]
is_palindromic_sum(1234)

```

```python
#14. count_ascending_digits(n): Count if digits of number are in ascending order. Example: 123 →
 #ascending → True, 321 → False. Hint: Compare each digit with the next.
def count_ascending_digits(n):
    s = str(n)
    for i in range(len(s) - 1):
        if s[i] > s[i + 1]:
            return False
    return True
count_ascending_digits(123)

```

```python
# 15. difference_of_extreme_digits(n): Return difference between largest and smallest digits.
 #Example: 58231 → max=8, min=1 → diff=7. Hint: Convert digits to list or string for max(), min().
def difference_of_extreme_digits(n):
    n = str(n)
    return int(max(n)) - int(min(n))
difference_of_extreme_digits(58231)
```

```python
 #16. sum_of_pentagonal_numbers(n): Return sum of first n pentagonal numbers. Example: n=3
# → (1*(3*1-1)/2)+(2*(3*2-1)/2)+(3*(3*3-1)/2)=1+5+12=18. Hint: Apply formula (3n² - n)/2 for each term.
def sum_of_pentagonal_numbers(n):
    sum = 0
    for i in range(1, n + 1):
        pent = (3 * i * i - i) // 2
        sum += pent
    return sum
sum_of_pentagonal_numbers(3)
```

```python
#17. sum_of_odd_series(n): Return sum of first n odd cubes (1³+3³+5³...). Example: n=3 →
# 1³+3³+5³=1+27+125=153. Hint: Use range(1, 2n, 2) and sum(i**3).
def sum_of_odd_series(n):
    sum = 0
    for i in range(1, 2 * n, 2):
        sum += i ** 3
    return sum
sum_of_odd_series(3)
```

```python
 #18. sum_of_even_series(n): Return sum of first n even squares (2²+4²+6²...). Example: n=3 →
# 2²+4²+6²=4+16+36=56. Hint: Use range(2, 2n+1, 2) for even numbers.
def sum_of_even_series(n):
    sum = 0
    for i in range(2, 2 * n + 1, 2):
        sum += i * i
    return sum
sum_of_even_series(3)
```

```python
#19. sum_of_factor_pairs(n): Return sum of all unique factor pairs (a,b) such that a*b=n. Example:
 #n=12 → pairs=(1,12),(2,6),(3,4) → sum=28. Hint: Loop through 1 to √n and sum both factors.
def sum_of_factor_pairs(n):
    sum = 0
    i = 1
    while i * i <= n:
        if n % i == 0:
            sum += i + (n // i)
        i += 1
    return sum
sum_of_factor_pairs(12)
```

```python
 #20. digit_difference_series(n): Return total of |difference between adjacent digits|. Example: 1234
# → |1-2|+|2-3|+|3-4|=3. Hint: Convert to string and use abs(int(a)-int(b)) for adjacent pairs.
# The Coders Orbit
def digit_difference_series(n):
    s = str(n)
    total = 0
    for i in range(len(s) - 1):
        total += abs(int(s[i]) - int(s[i + 1]))
    return total
digit_difference_series(1234)
```
